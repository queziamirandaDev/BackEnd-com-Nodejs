```python
df_vendas = pd.read_csv('base_vendas.csv', parse_dates=['dt_venda', 'dt_entrega']) # converter a coluna objects para data
```

```python
df_vendas.info()
```

```python
nome_colunas = df_vendas.columns
```

```python
nome_colunas = list(nome.lower().strip().replace(' ','_') for nome in nome_colunas)

print(nome_colunas)
```

```python
df_vendas.columns = nome_colunas
df_vendas.columns
```

##Modificando Tipos De Dados

```python
df_vendas.matricula_funcionario = df_vendas.matricula_funcionario.astype(str)
```

```python
df_vendas.info()
```

```python
df_vendas = df_vendas.astype({
    'codigo_produto' : 'str',
    'nome_loja': 'str'
})
```

```python
df_vendas.info()
```

```python
# change virgula for pontin to converter the type of dado to float

df_vendas['preco_custo'] = df_vendas['preco_custo'].str.replace(',','.')
df_vendas['valor_unitario'] = df_vendas['valor_unitario'].str.replace(',', '.')
```

```python
df_vendas[['preco_custo', 'valor_unitario']]
```

```python
# convertendo em float

df_vendas['preco_custo'] = df_vendas['preco_custo'].astype(float).replace(',','.')
df_vendas['valor_unitario'] = df_vendas['valor_unitario'].astype(float).replace(',','.')
```

```python
df_vendas[['preco_custo', 'valor_unitario']]
```

```python
df_vendas.info()
```

```python
df_vendas['dt_entrega'] = pd.to_datetime(df_vendas['dt_entrega'],
                                         format='%d/%m/%Y %H:%M')
```
