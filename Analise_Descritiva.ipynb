```python
## Renomeando as colunas

df_complementares = df_complementares.rename(columns={
    'Data':'dt_venda',
    'Loja':'nome_loja',
    'Produto':'descricao_produto',
    'Preço Unitário': 'valor_unitario',
    'Valor Final': 'valor_final',
    'Quantidade':'quantidade'
})
```

```python
df_complementares.columns
```

```python
df_complementares2= pd.concat([df_vendas,df_complementares], axis=0, join='inner')
#axis=0, join='outer') #tras tudo oq tem na tabela -- inner tras so oq elas tem em comum
df_complementares2.head()
```

```python
df_complementares2 = df_complementares.marge()
```

```python
df_join.head()
```

```python
df_complementares.head()
```

```python
df_complementares2.head()
```

```python
df_complementares2['dt_venda'] = pd.to_datetime(df_complementares2['dt_venda'])
```

```python
df_complementares2.info()
```

```python
df_complementares2['ano'] = df_complementares2['dt_venda'].dt.year
df_complementares2['mes'] = df_complementares2['dt_venda'].dt.month

```

```python
df_total = df_complementares2.merge(df_join ,on=['ano','mes'], how='inner')
```

```python
df_total.head()
```
